```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
text ="HI fuse hole shiyat. yo mannnn"
text1 = word_tokenize(text)
print(text1)
```

```python
text2 = sent_tokenize(text)
print(text2)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
input = "I knew it biyacth"
doc = nlp(input)
```

```python
for token in doc:
  print(token)
```

```python
for sent in doc.sents:
  print(sent)
```
