```python
pattern = r'\s'
data['char_count']= data['twitts'].apply(lambda x: len(re.sub(pattern, '', x)))
```

```python
data['word_count']= data['twitts'].apply(lambda x: len(x.split()))
```

```python
data['avg_word_length']=data['char_count']/data['word_count']
```

```python
data['hash_tags_count']=data['twitts'].apply(lambda x: len(re.findall(r'#\w+',x)))
```

```python
pattern = r'@\w+'
data['twitts'].apply(lambda x: re.sub(pattern, '', x))
```

```python
pattern = r'[^\w\s]'
data['twitts'].apply(lambda x: re.sub(pattern, '', x))
```

```python
pattern= r'\w+@\w+.\w+'
```

```python
data['email'] = data['twitts'].apply(lambda x: re.findall(pattern=pattern, string=x))
data['email'].sample(5)
```
